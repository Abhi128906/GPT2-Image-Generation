```python

# IMAGE GENERATION USING STABLE DIFFUSION
# PRODIGY INFOTECH TASK


# STEP 1: Install Libraries
!pip install diffusers transformers accelerate torch -q

# STEP 2: Import Libraries
from diffusers import StableDiffusionPipeline
import torch
import matplotlib.pyplot as plt

# STEP 3: Load Pretrained Model
model_id = "runwayml/stable-diffusion-v1-5"

pipe = StableDiffusionPipeline.from_pretrained(
    model_id,
    torch_dtype=torch.float16
)

pipe = pipe.to("cuda")

print("Model Loaded Successfully!")

# STEP 4: Enter Text Prompt
prompt = "A roaring Lion on a huge Rock where a photographer is clicking the pictures of the lion"

# STEP 5: Generate Image
image = pipe(prompt).images[0]

print("Image Generated Successfully!")

# STEP 6: Display Image
plt.imshow(image)
plt.axis("off")
plt.show()

# STEP 7: Save Image
image.save("generated_image.png")

print("Image Saved Successfully!")
```
